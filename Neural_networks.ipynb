```python
data=read_csv('Alphabets_data.csv')
data
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
x=data.drop(columns=['letter']).values
y=data['letter'].values
```

```python
label_encoder=LabelEncoder()
y_encoded=label_encoder.fit_transform(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y_encoded,test_size=0.2,random_state=1)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
model=Sequential()
model.add(Dense(120,activation='relu',input_shape=(x_train.shape[1],)))
model.add(Dense(64,activation='relu'))
model.add(Dense(14,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='SGD',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history=model.fit(x_train,y_train,epochs=150,batch_size=32,validation_split=0.2)
```

```python
loss,accuracy=model.evaluate(x_test,y_test)
```
